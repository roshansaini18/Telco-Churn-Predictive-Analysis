```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize=(4,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
colors = ['#2ECC71', '#E74C3C']  # green = no churn, red = churn
plt.pie(gb['Churn'], labels=gb.index, autopct="%1.2f%%", startangle=90, colors=colors)
plt.title("Percentage of Churned Customers", fontsize=12, weight='bold')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="gender", data=df, hue="Churn", palette="Paired")
plt.title("Churn by Gender", fontsize=13, weight='bold')
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.legend(title="Churn")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

fig, ax = plt.subplots(figsize=(5,4))
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#2ECC71','#E74C3C'])

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width/2, y + height/2, f'{height:.1f}%', ha='center', va='center', fontsize=9, color='black')

plt.title('Churn by Senior Citizen (in %)', fontsize=13, weight='bold')
plt.xlabel('Senior Citizen')
plt.ylabel('Percentage')
plt.legend(title='Churn', bbox_to_anchor=(1,1))
sns.despine()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(x="tenure", data=df, bins=72, hue="Churn", palette='Set2', kde=True)
plt.title("Distribution of Tenure by Churn", fontsize=13, weight='bold')
plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Customers")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x="Contract", data=df, hue="Churn", palette="coolwarm")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churn by Contract Type", fontsize=13, weight='bold')
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.legend(title="Churn")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
df.columns.values
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))
axes = axes.flatten()

for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, hue=df["Churn"], ax=axes[i], palette="Set3")
    axes[i].set_title(f'{col}', fontsize=12, weight='bold')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')
    axes[i].legend(title='Churn', loc='upper right')

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle("Customer Services vs Churn", fontsize=14, weight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(y="PaymentMethod", hue="Churn", data=df, palette="coolwarm")
plt.title("Churned Customers by Payment Method", fontsize=13, weight='bold')
plt.xlabel("Number of Customers")
plt.ylabel("Payment Method")
plt.legend(title="Churn", loc='best')
sns.despine()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='tenure', y='MonthlyCharges', hue='Churn', data=df, alpha=0.7)
plt.title('Monthly Charges vs Tenure by Churn Status')
plt.xlabel('Tenure (Months)')
plt.ylabel('Monthly Charges')
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
numeric_df = df[['tenure', 'MonthlyCharges', 'TotalCharges']].copy()
numeric_df['Churn'] = df['Churn'].map({'Yes':1, 'No':0})

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Contract', hue='PaymentMethod', data=df, palette='Set3')
plt.title('Payment Method Distribution by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.legend(title='Payment Method', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='InternetService', y='MonthlyCharges', hue='Churn', data=df, palette='Set2')
plt.title('Monthly Charges by Internet Service and Churn')
plt.xlabel('Internet Service Type')
plt.ylabel('Monthly Charges')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.kdeplot(data=df, x='tenure', hue='Contract', fill=True)
plt.title('Tenure Distribution by Contract Type')
plt.xlabel('Tenure (Months)')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```
